```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('real_estate.csv', low_memory = False)
df.head(3)
```

```python
df.info()
```

```python
df.isnull().sum()
```

Calculate the average price of property according to the Residential Type and show it using a barplot.

```python
Residential_Type_group = df.groupby('Residential Type')
```

```python
data = Residential_Type_group['Sale Amount'].mean().reset_index()
```

```python
plt.figure(figsize = (10,8))
sns.barplot(data = data, x = 'Residential Type', y = 'Sale Amount')
plt.ylabel('Average Sale Amount')
plt.show()
```
